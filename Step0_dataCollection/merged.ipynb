```python
import pandas as pd
```

```python
health_xlsx = "County Health Rankings-Final.xlsx"
health_df = pd.read_excel(health_xlsx)
health_df = health_df.rename(columns={"overall county rank": "health rank"})
health_df = health_df.drop(["state", "county"], axis=1)
health_df
```

```python
density_csv = "pop_density_ranked.csv"
density_df = pd.read_csv(density_csv)
density_df = density_df.rename(columns={"county_fips": "fips"})
density_df = density_df.drop(["county_name", "state_id", "county_name"], axis=1)
density_df
```

```python
gender_csv = "Gender_Final_ranked.csv"
gender_df = pd.read_csv(gender_csv)
gender_df = gender_df.drop(["state", "county"], axis=1)
gender_df
```

```python
age_csv = "Age_Final_ranked.csv"
age_df = pd.read_csv(age_csv)
age_df = age_df.drop(["state", "county"], axis=1)
age_df = age_df.rename(columns={"over 70": "% over 70", "Rank": "Rank over 70", "Scale": "scale over 70"})
age_df
```

```python
allages_csv = "Age_Final.csv"
allages_df = pd.read_csv(allages_csv)
allages_df = allages_df.drop(["state", "county"], axis=1)
allages_df
```

```python
poverty_csv = "Poverty_ranked.csv"
poverty_df = pd.read_csv(poverty_csv)
poverty_df = poverty_df.rename(columns={"FIPStxt": "fips"})
poverty_df = poverty_df.drop(["Stabr", "Area_name"], axis=1)
poverty_df
```

```python
income_csv = "Median_income_ranked.csv"
income_df = pd.read_csv(income_csv)
income_df = income_df.rename(columns={"FIPStxt": "fips"})
income_df = income_df.drop(["Stabr", "Area_name"], axis=1)
income_df
```

```python
transportation_csv = "Transportation_May_4.csv"
transportation_df = pd.read_csv(transportation_csv)
transportation_df = transportation_df.rename(columns={"Percent": "% using transportation", "county_fips": "fips"})
transportation_df = transportation_df.drop(["county_name", "state_id"], axis=1)
transportation_df
```

```python
covid411_csv = "April_11_data_ranked.csv"
covid411_df = pd.read_csv(covid411_csv)
covid411_df = covid411_df.drop(["County", "State"], axis=1)
covid411_df = covid411_df.dropna(how='any')
covid411_df = covid411_df.rename(columns={"FIPS": "fips", "Cases_rank": "Apr11_cases_rank", 
                                          "Cases_scale": "Apr11_cases_scale", "Deaths_rank":"Apr11_deaths_rank", 
                                          "Deaths_scale": "Apr11_deaths_scale" })
covid411_df
```

```python
covid418_csv = "April_18_data_ranked.csv"
covid418_df = pd.read_csv(covid418_csv)
covid418_df = covid418_df.dropna(how='any')
covid418_df = covid418_df.rename(columns={"FIPS": "fips"})
covid418_df
```

```python
latlong_csv = "uszips.csv"
latlong_df = pd.read_csv(latlong_csv)
latlong_df
```

```python
merge123456789_df = pd.merge(merge12345678_df, covid411_df, how='left', on = ['fips'])
merge123456789_df
```
